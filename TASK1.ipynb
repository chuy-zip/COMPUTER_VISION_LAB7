## Pregunta 1.1

El médico coordinador del proyecto le presenta la siguiente situación: "Tenemos 800 radiografías etiquetadas por nuestros radiólogos. Un colega me dijo que con tan pocos datos el modelo va a memorizar todo y no va a servir para nada en producción."

Como ingeniero de IA a cargo, usted decide aplicar Data Augmentation como parte de la solución. Sin embargo, el médico le pregunta: "¿No estamos inventando datos falsos que podrían confundir al modelo?"

Con esto en mente, responda las siguientes preguntas en su reporte:

1. Explíquele al médico, en términos que él pueda entender, qué es el Data Augmentation y por qué las imágenes generadas no son datos falsos. Use la analogía que considere más apropiada.
2. En el contexto específico de radiografías de tórax, proponga tres transformaciones de Data Augmentation que serían válidas y justifique cada una. Luego, identifique una transformación que no debería aplicarse en este dominio médico y explique por qué podría comprometer la integridad diagnóstica del modelo.
3. ¿Es el Data Augmentation suficiente por sí solo para garantizar que el modelo generalice bien? Argumente su posición considerando otras variables del proceso de entrenamiento.

---

Respuesta: 

1. El data augmentation no es generación de datos aleatorios o falsos, es tomar las imagenes que ya se tienen y realizarles pequeños ajustes que no influyen en el significado de la imagen original. Por ejemplo, si uno tiene una radiografía impresa uno puede tomarle una foto y subirle el brillo a la caja de luz, rotarla un poco, acercarla y/o alejarla, pero esto no cambia la imagen original y su contenido como tal. Solo son distintas formas de ver la misma radiografía. El concepto es el mismo para una IA, tener una imagen con modificaciones ligeras ayuda a tener más variabilidad en las imagenes para que la IA aprenda más patrones y no funcione solo con imágenes "perfectas". No son datos falsos, es más como generar perspectivas y alternativas de ver las imágenes

2. 3 formas validas de data aumentation en este caso podrían ser justo las que se le pueden dar de ejemplo al doctor como explicación:
* En el primer caso se puede ajustar cosas relacionadas a la apariencia como el brillo y contraste. Las imágenes de rayos X pueden no tener el mismo contraste de colores, brillo y demás aspectos que afectan como se ve una imagen.
* Rotaciones simples también pueden ser opciones de transformaciones, siempre y cuando no sea demasiado. Por ejemplo unos 5 o 10 grados es lo máximo que se podría manejar sin que la imagen comience a perder sentido. Lo que pasa es que por ejemplo para una radiografía una persona generalmente no se logra colocar completamentee derechos o alineados al sacar la radio.
* Por último un zoom ligero también podría ser aceptable, esto representaría cuando una persona esta más o menos cerca de la placa para poder tomar la radiografía. 

- El que ya no sería valido sería hacerl un espejo o flip a la imagen. Nuestro cuerpo no es simétrico, entonces tomar una imagen y darle la vuelta representaría algo erroneo y falso.

3. El data augmentation es una técnica muy eficiente para ayudar a un modelo a poder generalizar mejor, pero no es la única metodología para mejorar la generalización del modelo. POr si solo no está garantizado que la IA no tenga overfitting. El data augmentation ayuda bastante en aumentar la cantidad de datos y diversidad cuando la información es limitada, pero aún puede haber overfitting dependiendo de cuantos datos tengamos incluso luego del aumento. Para poder mitigar el overfitting también se deben tener en cuenta otros métodos o incluso tuneo de hiperparámetros. cosas como el learning rate, dropout o incluso el early stopping basado en la función de perdida.

Durante el entrenamiento del modelo de MediScan Guatemala, el equipo registra las siguientes curvas de pérdida (loss) al finalizar 25 épocas:

Respuesta:
4. Desde mi punto de vista el overfitting comeinza cuando se pasa de la época 10 a la 15. Se puede ver claramente que en ese punto la pérdida en entrenamiento sigue bajando, pero la pérdida en validación sube a partir de esa epoca al pasar de 0.38 a 0.42. Luego de eso la pérdida en validación sigue aumentando y en el caso de la perdida en entrenamiento llega a casi 0. Entonces claramente es mucho mejor con imagenes que conoce, pero con imágenes nuevas al final no es bueno clasificando.

5. Para evitar el overfitting se puede:
- Aplicar dropout es una opción, ya quee se "apagan" aleatoriamente cierto porcentaje de neuronas en la etapa de entrenamiento. Visto de una forma intuitiva lo que estamos evitando es que los pesos se adapten o tengan sesgo por una o varias neuronas en específico. Da mayot variabilidad y reduce la dependencia de una parte específica de la red neuronal. Puede que el loss en entrenamiento no sea tan bajo, pero se vería una mejora en la gráfica de pérdida de validación. Matemáticamente, tardaría más en reducir la pérdida en entrenamiento, pero eso también ayudaría y obligaría a que se exploren más caminos de la red.

- También se puede aplicar early stopping, aunque es un método un poco más simple puede ayudar a reducir el overfitting e incluso tiempo de entrenamiento innecesario. Este consiste en monitorear la métrica de validación durante el entrenamiento, de tal forma que que si el loss del a valifación deja de mejorar o comienza a empeorar luego de una cantidad de épocas seguidas, se detiene el entreamiento y se restauran los pesos del última mejor versión. Es muy importante mencionar que aunque es sencillo, es un método que no reduce el overfitting ded raíz y matemáticamente no afecta el cálculo de pesos, solo mitiga el overfitting al detener cuando deja de mejorar. En la gráfica de pérdida nos hubieramos quedado justo en la época 10.

6. Desde el punto de vista médico es bastante peligroso ya que estaríamos confiando en un modelo que puede dar muchos diagnósticos de falsos negativos. Ya que el modelo solo es bueno con las imágenes que conoce, con imágenes nuevas es muy probable que se confunda y que marqué a una persona con pneumonía como sana o incluso al revés. Pero el caso más catastrófico definitamentes es indetificar a alguien como sano cuando realmente no lo está, solo porque la imagen no coincidía con lo que el modelo tiene memorizado.
